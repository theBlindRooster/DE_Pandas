```python
import pandas as pd
import numpy as np

file1="raw_data/schools_complete.csv"
schools=pd.read_csv(file1)
schools.head()
```

```python
file2="raw_data/students_complete.csv"
students=pd.read_csv(file2)
students.head()
```

```python
totSchools = len(schools["name"].unique())
totStudents = len(students["name"].unique())
totBudget = schools["budget"].sum()
avgMath = round(students["math_score"].mean(),2)
avgRead = round(students["reading_score"].mean(),2)
passMath = round((((students[students["math_score"]>70].count()).math_score)/totStudents)*100,2)
passRead = round((((students[students["reading_score"]>70].count()).math_score)/totStudents)*100,2)
overallPass=(passMath+passRead)/2

print("District Summary")
print("-------------------")
print("Total Schools: "+str(totSchools))
print("Total Students: "+str(totStudents))
print("Total Budget: "+str(totBudget))
print("Average Math Score: "+str(avgMath))
print("Average Reading Score: "+str(avgRead))
print("% Passing Math: "+str(passMath))
print("% Passing Reading: "+str(passRead))
print("Overall Passing Rate: "+str(overallPass))
```

```python
schools = schools.rename(columns={"name":"school"})

grpStudents = students.groupby(["school"])

studentCount = grpStudents["Student ID"].count()

grpPassM = students[students["math_score"]>=70].groupby(["school"])
grpPassM = grpPassM['math_score'].count()

grpPassR = students[students["reading_score"]>=70].groupby(["school"])
grpPassR = grpPassR['reading_score'].count()

avgPassM = grpStudents['math_score'].mean()
avgPassR = grpStudents['reading_score'].mean()

schoolSum=pd.DataFrame({"Students":studentCount,"# Passing Math":grpPassM,"Avg Math Score":avgPassM,"# Passing Reading":grpPassR,"Avg Reading Score":avgPassR})

schoolSum = schoolSum.reset_index()

mergeTable=pd.merge(schoolSum,schools[["school","type","budget"]], on="school")
mergeTable["Budget Per Student"]=mergeTable["budget"]/mergeTable["Students"]
mergeTable["% Passing Math"]=round((mergeTable["# Passing Math"]/mergeTable["Students"])*100,2)
mergeTable["% Passing Reading"]=round((mergeTable["# Passing Reading"]/mergeTable["Students"])*100,2)
mergeTable["Overall Passing"]=round((mergeTable["% Passing Math"]+mergeTable["% Passing Reading"])/2,2)
mergeTable.head()
```

```python
topPerformers=mergeTable.sort_values(by=["Overall Passing"], ascending=False).head(5)
topPerformers
```

```python
bottomPerformers=mergeTable.sort_values(by=["Overall Passing"]).head(5)
bottomPerformers
```

```python
grpStdGrd = students.groupby(["school","grade"])
grpStdGrd[["math_score"]].mean().sort_values(by=["math_score"], ascending=False).head()
```

```python
perStudent = mergeTable[["school","Avg Math Score","Avg Reading Score","% Passing Math","% Passing Reading","Overall Passing","budget","Students"]]
perStudent["Per Student Spending"] = perStudent["budget"]/perStudent["Students"]
perStudent.sort_values(by=["Per Student Spending"], ascending=False)
```

```python
psBins=[578.0, 597.25, 616.5, 635.75, 655.0]
psLabels=["Low","Med-Low","Med-High","High"]

# Cut postTestScore and place the scores into bins
perStudent["Per Student Spending Rating"]=pd.cut(perStudent["Per Student Spending"], psBins, labels=psLabels)
perStudent.sort_values(by=["Overall Passing"], ascending=False)
```
